## **while loops**

Let’s move on to now understanding while loops. Unlike for loops, which typically stop after a fixed number of times, while loops will stop only when a specific condition is met.

What do you guys think will happen when I run this code?

Go ahead and write your own while loop.

```python
x=0
while x<10:
  x=x+1
  print(x)
print('done!')
```

In this example, x is the variable. x starts at 0 and increased by 1 each time the loop is run according to the formula x=x+1. Once 10 is reached, the condition to end the loop has been met, and the loop is finished. You will then see ‘done!’ printed.

The last code we ran was a for loop – this is called a while loop. Loops are useful because they can control our progress through our code; the ‘done!’ will not print until the loop has stopped running.

# **Part 5: Conditional Statements (If, Else)**

Just as we make descisions in real life with "If I do x, then y will happen" computers can also make decisions.
If statements allow us to control the path that a computer will take to execute a command. We can set up a condition to be met before running a block of code. Generally this expression is a Boolean (True or False) expression. Another way of thinking about it is basically you're telling the computer to ask a yes or no question.

In python, we need a colon after this condition and then we need to indent on the next line.


```python
feeling = input()

if feeling == "tired":
    print("Go back to bed")
```

What if we aren't tired though? This is where else and elif become useful.

```python
feeling = input()

if feeling == "tired":
  print("Go back to bed")
else:
  print("go get some coffee")
```

We can add in elif if we have another condition we want to check.

```python
feeling = input()

if feeling == "sick":
  print("Go back to bed")
elif feeling == "awake": #if the first statement was not true check another condition
  print("go make MJ coffee")
else:
  print("go get some coffee")
```

By the way, we can also use our membership operators in conditional statements:

# **Part 6: Loops**

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

With the for loop we can execute a set of statements, once for each item in a sequence.

```python
#example
hobbies = ["art", "photography", "reading"]
for a_hobby in hobbies:
  print(a_hobby)
```

```python
#another example
hobbies = ["art", "photography", "reading"]
for a_hobby in hobbies:
  print(a_hobby.upper())
```

```python
#Strings are iterable objects, they contain a sequence of characters
#Loop through the letters in the word "Onomatopoeia":

for letter in "Onomatopoeia":
  print(letter)
```

```python
#With the break statement we can stop the loop before it has looped through all
#the items

#Exit the loop when a_hobby is "photography":
hobbies = ["art", "photography", "reading"]
for a_hobby in hobbies:
  print(a_hobby)
  if a_hobby == "photography":
    break
```

```python
#Exit the loop when a_hobby is "photography",
#but this time the break comes before the print:

hobbies = ["art", "photography", "reading"]
for a_hobby in hobbies:
    if a_hobby == "photography":
        break
    print(a_hobby)
```

```python
#With the continue statement we can stop the current iteration of the loop,
#and continue with the next

#Do not print photography:
hobbies = ["art", "photography", "reading"]
for a_hobby in hobbies:
    if a_hobby == "photography":
        continue #skip this iteration
    else: print(a_hobby)
```

To loop through a set of code a specified number of times, we can use the range() function,

The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```python
#Using the range() function:
for x in range(6):
  print(x)
```



Note that range(6) is not the values of 0 to 6, but the values 0 to 5.

The range() function defaults to 0 as a starting value, however it is possible to specify the starting value by adding a parameter: range(2, 6), which means values from 2 to 6 (but not including 6):

```python
#Using the start parameter:
for x in range(2, 6):
  print(x)
```

The range() function defaults to increment the sequence by 1, however it is possible to specify the increment value by adding a third parameter: range(2, 30, 3):

```python
#Increment the sequence with 3 (default is 1):
#range(min, max, step size)
for x in range(2, 30, 3):
  print(x)
```

```python
#The else keyword in a for loop specifies a block of code to be executed when the loop is finished:
#Print all numbers from 0 to 5, and print a message when the loop has ended:
for x in range(6):
  print(x)
# else:
#   print("Finally finished!")
print("finally finished")
```

Note: The else block will NOT be executed if the loop is stopped by a break statement.

```python
#Break the loop when x is 3, and see what happens with the else block:
for x in range(6):
  if x == 3:
    break
  print(x)
else:
  print("Finally finished!")
```

A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop":

```python
#Print each adjective for every hobby:
adj = ["fun", "exciting", "boring"]
hobbies = ["art", "photography", "reading"]

for adjective in adj:
  for a_hobby in hobbies:
    print(adjective, a_hobby)
```

The While loop:
With the while loop we can execute a set of statements as long as a condition is true.

```python
#Print i as long as i is less than 6:
i = 0
while i < 10:
  print(i)
  i += 1 #helps i eventually meet the condition and stop the loop
```



Note: remember to increment i, or else the loop will continue forever.

We need to define an indexing variable, i, which we set to 0.

```python
#With the break statement we can stop the loop even if the while condition is true:
i = 0
while i < 10:
  print(i)
  if i == 4:
    break
  i += 1
```

```python
#With the continue statement we can stop the current iteration, and continue with the next:
#Continue to the next iteration if i is 4:
i = 0
while i < 10:
  i += 1
  if i == 4:
    continue #skip 4 and continue with loop
  print(i)
```

```python
#With the else statement we can run a block of code once when the condition
#no longer is true

#Print a message once the condition is false:
i = 0
while i < 10:
  print(i)
  i += 1
else:
  print("i is no longer less than 10")
```
